```python
import matplotlib.pyplot as plt
```

```python
from numpy.random import randn
import numpy as np
```

```python
plt.plot(randn(30).cumsum(), color='g', linestyle='dashed', marker='o')
```

```python
fig = plt.figure(); ax = fig.add_subplot(1, 1, 1)
ax.plot(randn(1000).cumsum())
```

# Adding ticks and labels

```python
fig = plt.figure(); ax = fig.add_subplot(1, 1, 1)
ax.plot(randn(1000).cumsum())
ticks = ax.set_xticks([0, 250, 500, 750, 1000])
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=30, fontsize='small')
ax.set_title('My first matplotlib plot')
ax.set_xlabel('Stages')
```

```python
fig = plt.figure(); ax = fig.add_subplot(1, 1, 1)
ax.plot(randn(1000).cumsum(), 'r', label='one')
ax.plot(randn(1000).cumsum(), 'g--', label='two')
ax.plot(randn(1000).cumsum(), 'b.', label='three')
ax.legend(loc='best')
```

# Line plot

```python
from pandas import Series
import pandas as pd
s = Series(np.random.randn(10).cumsum(), index=np.arange(0, 100, 10))
s.plot()
```

```python
from pandas import DataFrame
df = DataFrame(np.random.randn(10, 4).cumsum(0),
columns=['A', 'B', 'C', 'D'],
index=np.arange(0, 100, 10))
df.plot()
```

```python
import pandas as pd
tips = pd.read_csv('tips.csv')
tips
```

```python
day = Series(tips['day'])
x = Series(tips['tip']).round()
tip = x.astype('int32')
party_counts = pd.crosstab(day, tip)
party_counts
```

```python
party_counts = party_counts.loc[:,1:5]
party_counts
```

```python
party_counts.plot(kind='bar', stacked=True,color= ['Red','Orange','Blue','Green', 'Purple'])
plt.savefig('tip.png')
```

# Pie Chart

```python
labels = ['Orange','Apple','Mango','Cherry']
sizes = [25,30,45,65]
colors = ['Orange','Red','Yellow','Blue']
plt.pie(sizes, labels= labels, colors = colors, autopct = '%1.1f%%')
plt.title('Fruit Distribution',fontsize = 20,weight='bold')
plt.axis('equal')
plt.show()
```

# Histogram

```python
tips['tip_pct'] = tips['tip'] / tips['total_bill']
x = tips['tip_pct'].hist(bins=50)
x.set_xlabel('Tip Percentage')
x.set_ylabel('Frequency')
x.set_title('Tip Percentage Histogram')
```

# Scatter plot

```python
macro = pd.read_csv('macrodata.csv')
data = macro[['cpi', 'm1', 'tbilrate', 'unemp']]
trans_data = np.log(data).diff().dropna()
trans_data[-5:]
```

```python
plt.scatter(trans_data['m1'], trans_data['unemp'])
plt.title('Changes in log %s vs. log %s' % ('m1', 'unemp'))
plt.savefig('scatter_plot.png', dpi=150, bbox_inches='tight')
```
